```python
import tkinter as tk
from tkinter import messagebox
import csv

def check_employee():
    name = name_entry.get()
    department = dept_entry.get()

    found = False
    members = []

    with open("GIG-logistics(in).csv", "r") as file:
        reader = csv.DictReader(file)

        for row in reader:
            if row["Department"].lower() == department.lower():
                members.append(row["Name"])

            if row["Name"].lower() == name.lower() and row["Department"].lower() == department.lower():
                found = True

    result_box.delete("1.0", tk.END)

    if found:
        result_box.insert(tk.END, "Employee exists\n")
        result_box.insert(tk.END, "Welcome " + name + "\n\n")
        result_box.insert(tk.END, "Members in " + department + " department:\n")

        for member in members:
            if member.lower() != name.lower():
                result_box.insert(tk.END, member + "\n")
    else:
        messagebox.showinfo("Result", "Employee does not exist")

root = tk.Tk()
root.title("GIG Logistics")
root.geometry("500x400")

tk.Label(root, text="Enter Name").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Enter Department").pack()
dept_entry = tk.Entry(root)
dept_entry.pack()

tk.Button(root, text="Check Employee", command=check_employee).pack()

result_box = tk.Text(root, height=15, width=50)
result_box.pack()

root.mainloop()
```
