```python
# Simple Interest

P = float(input("Enter principal: "))
R = float(input("Enter rate (%): "))
T = float(input("Enter time: "))

A = P * (1 + (R * T) / 100)

print("Amount =", A)
```

```python
# compound interest

P = float(input("enter principal: "))
R = float(input("enter rate (%) "))
T = float(input("enter time: "))
n = float(input("enter no of compounded times : "))


A = P * (1 + R/n) ** (n*T)

print("Amount =", A)
```

```python
#Annuity plan

# Annuity Plan

PMT = float(input("Enter payment each period: "))
R = float(input("Enter interest rate: "))
n = int(input("Enter number of payments per year: "))
t = float(input("Enter number of years: "))

A = PMT * (((1 + R/n) ** (n*t) - 1) / (R/n))

print("Future Value =", A)


```
